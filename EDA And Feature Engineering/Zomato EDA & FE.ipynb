#### Here is a brief description of each data type (Dtype):

- ***object*** is the most general data type in Python. It can be used to store any type of data, including strings, numbers, lists, and dictionaries.
- ***int*** is the data type for integers. Integers are whole numbers, such as 1, 2, 3, and so on.
- ***float*** is the data type for floating point numbers. Floating point numbers are numbers with decimal places, such as 1.23, 3.14, and so on.
- ***complex*** is the data type for complex numbers. Complex numbers are numbers of the form a + bi, where a and b are real numbers and i is the imaginary unit.
- ***bool*** is the data type for Boolean values. Boolean values are either True or False.
- ***datetime*** is the data type for dates and times. datetime objects can represent dates, times, or both.
- ***timedelta*** is the data type for time intervals. timedelta objects can represent periods of time, such as days, hours, minutes, and seconds.
- ***str*** is the data type for text strings. str objects can represent any sequence of characters.
- ***bytes*** is the data type for binary data. bytes objects can represent any sequence of bytes.
- ***array*** is the data type for arrays. arrays are sequences of objects of the same type.
- ***None*** is the null value. None is used to represent the absence of a value.

```python
# using heatmap 
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

- ***matplotlib.rcParams['figure.figsize']=(12,6):*** This line sets the default figure size for matplotlib plots to be 12 inches in width and 6 inches in height. 
- ***df.isnull():*** The DataFrame representing missing values.
- ***yticklabels=False:*** This disables displaying y-axis tick labels in the heatmap. This can be useful to save space when the y-axis labels are not necessary.
- ***cbar=False:*** This disables the colorbar, which is a separate bar showing the color scale for the heatmap values. Disabling the colorbar can be useful when the scale is not needed or when it would take up too much space.
- ***cmap='viridis':*** This sets the color map for the heatmap. In this case, the 'viridis' colormap is used, which is a perceptually uniform and visually appealing colormap that goes from yellow to green to blue.
